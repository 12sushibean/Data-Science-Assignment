Introduction

GitHub, Jupyter Notebook and RStudio are pivotal tools in the data science ecosystem, each serving distinct yet complementary roles. GitHub acts as a collaborative platform for version control and project management, fostering teamwork and code sharing. Jupyter Notebook provides an interactive environment for creating and sharing documents containing live code, visualizations and explantory text, facilitating reproducible research and data exploration. Meanwhile, RStudio offers an integrated development environment tailored for R programming, equipped with tools for data analysis, visualization and statiscal modeling, caatering to the specific needs of R users. Together, these tools form a powerful toolkit for data scientists, enabling efficient collaboration, interactive analysis and seamless workflow management.
